```python
df = pd.read_csv('dataset/Vizari_data.csv')
num_rows_to_select = 100
offset = 6

# Create a list of row indices to select
indices_to_select = list(range(0, num_rows_to_select * offset, offset))

# Use iloc to select the rows based on the specified indices
df = df.iloc[indices_to_select]
```

```python
dataJson = df.to_json(orient='records')
print(dataJson)
#writing to json file
with open('data2.json', 'w') as f:
    f.write(dataJson)
```
